###NumPy (Array Operations)

```python
#1
import numpy as np
import pandas as pd
from random import randint
# Create a 5×5 array with random integers from 1 to 50
arr = np.random.randint(1,51,size =(5,5))
print(arr)

## Find the row with the maximum sum
row_sum = arr.sum(axis=1)
max_row_index =  np.argmax(row_sum)

print('\nRow Sums: ')
print(row_sum)

print("\nRow with the maximum sum:")
print(arr[max_row_index])

print("Index of the row:", max_row_index)

# Calculate the mean of the array
mean_value = arr.mean()
print("\nMean of the array:", mean_value)

# Replace elements greater than the mean with 0
modified_arr = arr.copy()
modified_arr[modified_arr > mean_value] = 0

print("\nModified Array:")
print(modified_arr)
```

```python
#2
## Create an array from 1 to 20
arr = np.arange(1,21)

print("Original Array:")
print(arr)

# Select every 3rd element starting from the 2nd element
new_arr = arr[1::3]
print("\nNew Array:")
print(new_arr)
```

###Pandas (Data Handling)

```python
#3
iris_df = pd.read_excel('/content/iris.xls')
iris_df.head(10)
```

```python
iris_df['Classification'].value_counts()
```

```python
iris_df.describe()
```

```python
iris_df.info()
```

```python
# Filter rows where sepal_length > median and species is 'versicolor'
median_sepal = iris_df['SL'].median()
filtered_df = iris_df[
    (iris_df['SL']>median_sepal) &
    (iris_df['Classification']=='Iris-versicolor')]

print("\nFiltered Rows:")
print(filtered_df)

# Group by species and compute mean petal_length
mean_petal = iris_df.groupby('Classification')['PL'].mean()

print("\nMean Petal Length by Species:")
print(mean_petal)
```

```python
#4
iris_df['petal_ratio']= iris_df['PL'] /iris_df['PW']
```

```python
# Find the top 5 rows with the highest petal_ratio
top_5 = iris_df.nlargest(5,'petal_ratio')
print("\nTop 5 rows with the highest petal_ratio:")
print(top_5)
```

###Data Visualization


```python
#5
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
plt.hist(iris_df['SL'],bins=15)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Plot each species separately
for species in iris_df['Classification'].unique():
     subset = iris_df[iris_df['Classification']==species]
     plt.scatter(
         subset['SL'],
         subset['PL'],
         label = species
     )

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend(title="Species")
plt.show()
```

```python
#6
sns.pairplot(iris_df, hue='Classification')
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.boxplot(
    x="Classification",
    y="PL",
    data= iris_df
)

plt.title("Petal Width Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width")

plt.show()
```
