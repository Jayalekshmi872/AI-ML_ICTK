```python
#1. Simple ATM Calculator
balance_amt = 5000
withdraw_amt = 1500
sufficient_balance = 500
if balance_amt > withdraw_amt:
  balance_amt-= withdraw_amt
  print(f'Transaction successful. Remaining balance: {balance_amt}')
else:
  print(f'No sufficient balance in your account')
```

```python
#2. Electricity Bill Generator
unit_consumed = int(input('Enter Units consumed: '))
bill =0

if unit_consumed <= 100: #Units less than or equal to 100
  bill= unit_consumed*5
elif unit_consumed<=200: #Units between 101 and 200
  bill = (100 *5)+((unit_consumed-100)*7)
else:
  bill= (100*5)+(100*7)+((unit_consumed-200)*10)

print("Bill amount = ₹", bill)

```

```python
#3. Temperature Converter

temp = float(input('Enter temperature: '))
while True:

  unit=input('Enter unit (C/F): ')
  if unit=="C":
    fah = (temp*9/5)+32
    print(f'Temperature in Fahrenheit: {fah:.2f} ℉')
    break
  elif unit=='F':
    cel =(temp-32)*5/9
    print(f'Temperature in Celsius:{cel:.2f} ℃')
    break
  else:
    print(input('Invalid unit! Please enter C or F.'))
```

```python
#4. Mobile Recharge Validity
from datetime import datetime, timedelta
today = datetime.strptime('2025-07-30','%Y-%m-%d')
Recharge_date= datetime.strptime('2025-07-01','%Y-%m-%d')
exp_date = Recharge_date+timedelta(days=28)
if today <= exp_date:
  print(f'Enjoy your recharge until {exp_date.date()}')
else:
  print('Plan expired. Recharge again. ')
```

```python
#5. Bus Seat Reservation System:
Booked = ['Amit', 'Sara', 'John','unni','sneha','niru','alli']
Cancel = 'Sara'
New = 'Neha'
Position = 1
# Remove cancelled passenger
if Cancel in Booked:
  Booked.remove(Cancel)
# Insert new passenger
Booked.insert(Position, New)
print(Booked)
```

```python
#6. Sum of even numbers
total =0
num=int(input('Enter a number: '))

for i in range(1,num+1):
   if i%2 == 0:
    total+=i
print('Sum of even numbers:', total)
```

```python
#7. Dictionary – Student Marks Summary
students ={}
for i in range(3):
  name =input('Enter name: ')
  marks = int(input('Enter marks: '))
  students[name]=marks
avg = sum(students.values())/len(students)
print('Average Marks: ',avg)
```

```python
#8. Dictionary – Word Frequency

text= input('enter text: ')

#print(text.lower().split())
word_count = {}
for word in text.lower().split():
  if word in word_count:
    word_count[word]+=1
  else:
    word_count[word]=1
print('word counts are: ',word_count)
```

```python
#8. Functions – Check Prime
prime =int(input('enter a number: '))
if prime/prime==0 or prime/1 ==0:
  print(f'{prime} is a prime number')
else:
  print(f'{prime} is not a prime number')
```

```python
#9. Vowel Counter
user_vowel = input('enter a string: ')
num_vowel =0
vowels =['a','e','i','o','u']
for char in user_vowel.lower():
  if char in vowels:
   num_vowel+=1
print(f'Number of vowels: {num_vowel} ')

```

```python
#10. Count Divisible Numbers
N = int(input('Enter range limit: '))
D = int(input('Enter divisor: '))
count=0
for i in range(1,N+1):
  if i%D==0:
    count+=1
print(f'count of numbers divisible by {D}: {count} ')
```
